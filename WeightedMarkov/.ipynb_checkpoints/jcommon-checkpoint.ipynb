```python
def readFile(file):
    with open(file, "rb") as f:
        c = f.read().decode().replace('\r\n', '\n')
        return c;

#c = readFile("PyUtils/ajax.html")
#display(HTML(c))

globalVAR=''
def GetFormValue(n='#graphtext' ):
    global globalVAR
    var='globalVAR'
    cmd='''GetFormValue('{}','{}')'''.format(n,var)
    display(Javascript(cmd))
    return globalVAR;

def GetJValue(jVar='edt1' ):
    global globalVAR
    var='globalVAR'
    cmd='''<script>SetPyVarValue({},'globalVAR') </script>'''.format(jVar)
    display(HTML(cmd))
    ret = globals()['globalVAR']
    #print("===> ", ret)
    return ret;


def GetDFTable(htmlTable, justLikeDF=None, copyRowsToDF=True):
    tJ = GetJValue("TableToJSON('{}')".format(htmlTable))
    
    tab=json.loads(tJ)
    table=[]

    for i, c in enumerate(tab):
        g =[g.strip() for g in c[0].split('\n')]
        g = g[1:-1]
        table.append(g)
        #print (i, len(g), g )

    dft = pd.DataFrame(table[1:], columns=table[0])
    idx=dft['']
    dft.drop('',axis=1, inplace=True)
    dft.index=idx
    
    if ( justLikeDF is None):
        return dft;
    
    dft.index = dft.index.astype(dff.index)

    for i,dt in enumerate(dft.dtypes):
        if ( dt != dff.dtypes[i]):
            #print ("Chaging type: ", i, dt, dff.dtypes[i])
            dft[dft.columns[i]] = dft[dft.columns[i]].astype(dff.dtypes[i])

    if ( copyRowsToDF):
        for i, r in dft.iterrows():
            dff.loc[i:i,:] = dft.loc[i:i,:]
    
    return dft
```
